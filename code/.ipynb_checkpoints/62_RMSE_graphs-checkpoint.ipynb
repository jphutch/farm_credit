# RMSE Graphs

This notebook uses data created in "60_R_Analysis.ipynb" and  creates Figure B1

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

with open("data_dir.txt") as f:
    data_dir = f.read()
```

```python
cropval = pd.read_csv(data_dir+"clean_data/crop_rmses.csv").iloc[:,1:].T

cropval.columns = ["M1","M2","M3","M4","M5","M6"]

sns.boxplot(data=cropval, palette="viridis")
plt.xlabel("Models")
plt.ylabel("RMSE")
plt.savefig("Crop_RMSEs.pdf",bbox_inches="tight",dpi=250)
```

```python
corn = pd.read_csv(data_dir+"clean_data/corn_rmses.csv").iloc[:,1:].T

corn.columns = ["M1","M2","M3","M4","M5","M6"]

sns.boxplot(data=corn, palette="viridis")
plt.xlabel("Models")
plt.ylabel("RMSE")
plt.savefig("../figs/Crop_RMSEs.pdf",bbox_inches="tight",dpi=250)
```

```python
tractor = pd.read_csv(data_dir+"clean_data/tractor_rmses.csv").iloc[:,1:].T

tractor.columns = ["M1","M2","M3","M4","M5","M6"]

sns.boxplot(data=tractor, palette="viridis")
plt.xlabel("Models")
plt.ylabel("RMSE")
```

## Figure B1

```python
f,a = plt.subplots(1,3,figsize=(15,5))

sns.boxplot(data=corn, palette="viridis",ax=a[0])
a[0].set_title("Corn Yield")
# a[0].set_xlabel("Models")
a[0].set_ylabel("RMSE")

sns.boxplot(data=cropval, palette="viridis",ax=a[1])
a[1].set_title("Crop Value per Acre")
# a[1].set_xlabel("Models")
a[1].set_ylabel("RMSE")

sns.boxplot(data=tractor, palette="viridis",ax=a[2])
a[2].set_title("Tractors per Farm")
# a[1].set_xlabel("Models")
a[2].set_ylabel("RMSE")

plt.savefig("../figs/Model_RMSEs.pdf",bbox_inches="tight",dpi=250)
```
