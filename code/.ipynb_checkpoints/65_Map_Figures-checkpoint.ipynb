# Maps

In this notebook, all of the Figures containing maps are created.

### Import line

```python
import geopandas as gp
import pandas as pd
import matplotlib.pyplot as plt
from shapely.geometry import MultiPolygon
from simpledbf import Dbf5
import matplotlib

%matplotlib inline

with open("data_dir.txt") as f:
    data_dir = f.read()
```

### Reading in Data

```python
# County shapefile
counties = gp.read_file(data_dir + 'spatial_data/tmean_total.shp')
counties = counties[counties.year==1920]
counties['FIPS'] = counties['FIPS'].astype(int)

# Get rid of these two extinct counties
# counties = counties.loc[~((counties.ICPSRNAM=='MILTON')&(counties.STATENAM=='Georgia'))]
# counties = counties.loc[~((counties.ICPSRNAM=='CAMPBELL')&(counties.STATENAM=='Georgia'))]

# List of each counties in PCA coverage area.
coverage = pd.read_csv(data_dir+ 'pca_data/county_list.csv')
coverage['FIPS_PCA'] = coverage['pca_id'].apply(lambda x: x.split("_")[0]).astype(int)

# The points for each PCA.
PCAs = gp.read_file(data_dir + "pca_data/pca_eqarea.shp")
PCAs = PCAs.rename(columns={"cityid":"pca_id",\
                            "pca_citi_1":"state",\
                            "pca_citi_2":"county",\
                            "pca_citi_3":"city"})

PCAs.loc[PCAs['pca_id'] =="48207_stam","pca_id"] = "48253_stam" 
PCAs.loc[PCAs['pca_id'] =="48207_stam","FIPS"] = 48253 
coverage['pca_id'] = coverage['pca_id'].replace("48207_stam","48253_stam")

# Filling in some missing IDs.
missing_ids = ["24011_dent","34011_brid","34005_moor","34023_newb","34027_morr","42091_norr","42077_alle"]
PCAs.loc[:6,"pca_id"] = missing_ids

PCAs['FIPS'] = (PCAs['SFIPS']+PCAs['CFIPS'])

# Missing one city, correcting here by adding it in.
city_file = data_dir + "pca_data/ci08au12.shp"
Cities = gp.read_file(city_file)
Cities = Cities.set_crs(epsg=4326)
Cities = Cities.to_crs(epsg=2163)
Cities = Cities.rename(columns={"cityid":"pca_id"})

pcalist = pd.read_csv(data_dir + "pca_data/pca_list.csv")
pca_cities = Cities.merge(pcalist,on="pca_id",how='right')
bennet = pca_cities[pca_cities['city'] =="Bennettsville"].drop_duplicates()
PCAs = pd.concat([PCAs,bennet],axis=0)

# PCA Districts
pca_districts = pd.read_csv(data_dir+'clean_data/fcs_final.csv',usecols=['FIPS','pca_district','pca_district_groups']).drop_duplicates()
PCA_districts = counties.merge(pca_districts,on="FIPS",how="left").dissolve(by="pca_district")
district_area_geom = PCA_districts.pop("geometry")
district_area_geom = district_area_geom.apply(lambda x: list(x) if isinstance(x, MultiPolygon) else x).explode()
PCA_districts = PCA_districts.join(district_area_geom, how='inner')

# States
states = counties.dissolve(by="STATENAM")
state_area_geom = states.pop("geometry")
state_area_geom = state_area_geom.apply(lambda x: list(x) if isinstance(x, MultiPolygon) else x).explode()
states = states.join(state_area_geom, how='inner')

# Finding whether the PCA is in the county.
counties = counties.merge(coverage[['FIPS_PCA']].drop_duplicates(),left_on="FIPS",right_on="FIPS_PCA",how='left')
counties['PCA'] = (~pd.isnull(counties.FIPS_PCA)).astype(int)
counties.drop("FIPS_PCA",axis=1,inplace=True)

# Measure latitude and longitude coordinates (in WGS 84)
counties['lat'] = counties.to_crs("EPSG:4326").centroid.x
counties['lon'] = counties.to_crs("EPSG:4326").centroid.y

# Merging PCA id in.
counties = counties.merge(coverage[['pca_id','FIPS']].drop_duplicates(),how='left')

# Coverage areas shapefile
coverage_areas = counties.dissolve("pca_id")
area_geom = coverage_areas.pop("geometry")
area_geom = area_geom.apply(lambda x: list(x) if isinstance(x, MultiPolygon) else x).explode()
coverage_areas = coverage_areas.join(area_geom, how='inner')
```

### Figure 3: PCA Bin Distances by County

```python
pca_distances = pd.read_csv(data_dir+'clean_data/fcs_final.csv',usecols=['FIPS','pca_km'])

pca_distances = counties.merge(pca_distances,how='left')

cmap = matplotlib.cm.get_cmap("Blues",6)
pca_distances['pca_cat'] = pd.cut(pca_distances['pca_km'],[0,30,45,60,100,1000])
pca_distances['pca_cat_code'] = pca_distances.pca_cat.cat.codes 

f,ax = plt.subplots(figsize=(20,40))
ax.axis('off')

pca_distances.plot(ax=ax,column="pca_cat_code",cmap=cmap,lw=0,edgecolor=None)
coverage_areas.geometry.exterior.plot(ax=ax,fc=None,edgecolor="black",alpha=.45,lw=.25)
PCA_districts.geometry.exterior.plot(ax=ax,fc=None,edgecolor='black',linewidth =1)
PCAs.plot(ax=ax,color="black",marker="*",label="Production Credit Association",lw=.1)

sm = plt.cm.ScalarMappable(cmap=cmap)
sm._A = []
#ax.set_title(title,fontsize=18)
cb = plt.colorbar(sm,fraction=0.02, pad=0.04,ax = ax);
cb.set_ticks([.1,.3,.5,.7,.9])
cb.set_ticklabels(['[0,30)','[30,45)','[45,60)','[60,100)',r'$\geq 100$'])
```

## Maps of Robustness Check Samples

### Figure 6: Robustness Check Samples

```python
samples = pd.read_csv(data_dir+"clean_data/samples.csv").iloc[:,1:]

sample_areas = counties.merge(samples,on="FIPS")
```

```python
from matplotlib.patches import Patch
from matplotlib.lines import Line2D

legend_elements = [Line2D([0], [0], marker='*',color='w', markerfacecolor='black', label='PCA', markersize=15),
                   Patch(facecolor='C3', edgecolor='black',label='States with One PCA'),
                   Patch(facecolor='C5', edgecolor='black',label='Western PCA District'),
                   Patch(facecolor='goldenrod', edgecolor='black',label='High Erosion Counties')]

f,a = plt.subplots(figsize=(20,40))
counties.plot(ax=a,color="lightsteelblue",lw=.5,edgecolor="black")
sample_areas[~sample_areas.restricted].plot(ax=a,color="C3",alpha=.2,label="One PCA States",legend=True)
sample_areas[~sample_areas.no_west].plot(ax=a,color="C5",alpha=.2)
sample_areas[(~pd.isnull(sample_areas.no_dust_bowl))&(~sample_areas.no_dust_bowl.dropna())].plot(ax=a,color="goldenrod",alpha=.2)
a.legend(handles=legend_elements,loc="lower left",fontsize=18)
PCAs.plot(ax=a,color="black",marker="*")
a.axis('off')
# plt.savefig("../../figs/sample_map.png",bbox_inches="tight",dpi=200)
```
