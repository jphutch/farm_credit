```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

with open("data_dir.txt") as f:
    data_dir = f.read()
    
pd.set_option("display.max_rows",500)
```

```python
df = pd.read_csv(data_dir+"clean_data/fcs_final.csv")

outcomes = ["cropval_pacr","corn_yield7","wheat_yield","labor_pacr","tract_pacr","fert_pacr"]

df['pca_cat'] = pd.cut(df['pca_km'],[0,30,45,60,100,500])
df['pca_cat'] = df.pca_cat.cat.rename_categories(["(0, 30]","(30, 45]","(45, 60]","(60, 100]",r"> 100"])
df['cropval_pacr'] = df['total_crop_val']/df['farm_area']
df['tract_pacr'] = df['equip_tractors']/df['farm_number']
df['labor_pacr'] = df['expend_labor']/df['farm_number']
df['fert_pacr'] = df['expend_fert']/df['farm_area']
df['avgfsize'] = df['farm_area']/df['farm_number']

df['outlier'] = \
((df.state_name =="Arizona") | \
 (df.state_name=="Utah") | \
 (df.state_name =="Nevada") | \
 (df.state_name =="Wyoming")| \
 (df.state_name =="New Mexico")|\
 (df.state_name =="Idaho")).astype(int)

def paralell_trends_discrete(outcome,title,ax):
    data = df.groupby(["PCA",'year'])[outcome].mean().reset_index().dropna()
    temp = {}
    for j in data.PCA.unique():
        temp[j] = data[data.PCA==j].set_index('year')[outcome]
    temp[0].plot(color="C0",marker="o",label="No PCA",ax=ax)
    temp[1].plot(color="C0",ls="--",marker="o",label="PCA",ax=ax)
    ax.set_xticks(temp[0].index)
    ax.axvline(1933,color='black',ls="--")
    ax.set_xlabel("")
    ax.set_title(title,fontsize=16)
    return ax

def paralell_trends_cat(outcome,title,ax):
#     fig, ax = plt.subplots()
    data = df.groupby(["pca_cat",'year'])[outcome].mean().reset_index().dropna()
    temp = {}
    linestyles = ["-","-.","--",":","-"]
    cmap = matplotlib.cm.get_cmap("viridis_r",9)
    for i,j in enumerate(data.pca_cat.unique()):
        temp[j] = data[data.pca_cat==j].set_index('year')[outcome]
        temp[j].plot(color=cmap(i+1),marker="o",label=str(j)+" km",ax=ax,ls=linestyles[i])
        ax.set_xticks(temp[j].index)
    ax.axvline(1933,color='black',ls="--")
#     plt.legend(loc =(1.05,.35))
    ax.set_xlabel("")
    ax.set_title(title,fontsize=16)
    return ax
```

### PCA in county or out

```python
f, axarr = plt.subplots(3,2,figsize=(16,18))
matplotlib.rcParams.update({'font.size': 14})

paralell_trends_discrete('farm_area',"Area in Farms (acres)",axarr[0,0])
paralell_trends_discrete('farm_number',"Number of Farms (#)",axarr[0,1])
paralell_trends_discrete('total_crop_val',"Crop Value ($)",axarr[1,0])
paralell_trends_discrete('avgfsize',"Average Farm Size",axarr[1,1])
paralell_trends_discrete('expend_fert',"Fertilizer ($)",axarr[2,0])
paralell_trends_discrete('equip_tractors',"Tractors (#)",axarr[2,1])

plt.legend(loc=[-.25,-.35],title = "PCA in County")
```

```python
f, axarr = plt.subplots(3,2,figsize=(16,18))
matplotlib.rcParams.update({'font.size': 14})

paralell_trends_discrete('corn_yield7',"Corn Yield (bu./acre)",axarr[0,0])
paralell_trends_discrete('wheat_yield',"Wheat Yield (bu./acre)",axarr[1,0])
paralell_trends_discrete('cropval_pacr',"Crop Value per Acre ($/acre)",axarr[2,0])
paralell_trends_discrete('fert_pacr',"Fertilizer per Acre ($/acre)",axarr[0,1])
paralell_trends_discrete('tract_pacr',"Tractors per Farm",axarr[1,1])
f.delaxes(axarr[2,1])
plt.legend(loc=[1.5,.25],title = "PCA in County")
```

### Distance to PCA

```python
f, axarr = plt.subplots(figsize=(8,6))
matplotlib.rcParams.update({'font.size': 14})
paralell_trends_cat('farm_area',"Area in Farms (acres)",axarr)
plt.legend()
plt.savefig('../../figs/trends_farmarea.pdf', bbox_inches='tight',dpi=250)
```
